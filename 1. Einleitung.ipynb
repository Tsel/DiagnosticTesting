##Was ist ein diagnostischer Test?

Ein diagnostischer Test ist ein binärer Klassifizierer, mit dem Objekte (Menschen, Tiere, Proben, ...) unter Verwendung eines Beobachtungswertes in die zwei Klassen "gesund (-)" und "krank (+)" eingeteilt werden.

Dabei ist zu beachten, dass die untersuchten Objekte in Wahrheit einer der beiden Klassen zugehörig sind, und der Beobachtungswert (das Ergebnis des diagnostischen Tests) als Hinweis auf diese "wahre" Zugehörigkeit anzusehen ist. Für jedes untersuchte Objekt liegen also zwei Merkmale (Attribute) vor, 1. der wahre Status, und 2. der Teststatus. Der wahre Status kann bekannt sein oder nicht und der Teststatus basierte auf der Einteilung der Objekte unter Verwendung des Beobachtungswertes (des Ergebnisses des diagnostischen Testes). 

Wir bezeichnen im Folgenden den wahren Status eines Objektes mit $D^+$ und $D^-$ und den Status aufgrund des Beobachtungswertes mit $T^+$ und $T^-$.

>Beispiel: Fieberthermometer als binärer Klassifizierer

>Das Fieberthermometer kann als binärer Klassifizierer aufgefasst werden. Hier wird die Körpertemperatur als Maß verwendet, um eine Einteilung in $T^+$ und $T^-$ vorzunehmen. Hierzu ist es notwendig einen Grenzwert festzulegen, über den die Transformation der kontinuierlichen Temperatur in den binären Status erfolgt.

>Die Basal-Körperkerntemperatur bei Schweinen liegt zwischen 39.3 und 39.9 Grad Celsius (wahrer Status $D^-$) und bei Fieber zwischen 40.5 und 41.1 Grad Celsius (wahrer Status $D^+$). In einen fiktiven Beispiel untersuchen wir 100 Schweine ohne Fieber und 100 Schweine mit Fieber und messen deren Körpertemperatur. Da wir keine echten Daten vorliegen haben, erzeugen wir entsprechende Daten aus den obigen Angaben. Die Messwerte der Körpertemperatur $c$ bei Tieren ohne Fieber bezeichnen wir mit $c^-$ und die Messwerte der Körpertempertur bei Tieren mit Fieber mit $c^+$. Bei der Erzeugung der Daten  gehen wir davon aus, dass die Körper - Temperaturen normalverteilt sind. Dann ist:  

$$

\mu_i, \sigma_i^2 = \begin{cases} (39.9+39.3)/2, ((39.9-39.3)/2)^2 \quad \text{ wenn } i \equiv -\\
(40.5+41.1)/2, ((41.1-40.5)/2)^2 \quad \text{ wenn } i \equiv +\\
\end{cases}
\qquad (1)
$$

Aber zurück zu der Frage: Was ist ein diagnostischer Test?
Ein diagnostischer Test ist eine Vorschrift, mit der wir Objekte zwei sich gegenseitig ausschliessenden Klassen zuordnen. 

Im obigen Beispiel haben wir Objekte und diese haben ein Attribut, nämlich die Temperatur. Dieses Attribut Temperatur soll nun verwendet werden, um die Tieren den Klassen $T^+$ und $T^-$ zuzuordnen. Dann brauchen wir eine Vorschrift, die dieses Attribut heranzieht, um die Klassifizierung vorzunehmen. Diese Vorschrift könnte wie folgt lauten:

> **Beispiel**

> Wenn die Temperatur eines Objektes größer als 39.9 Grad Celsium ist, dann ordne dieses Individuum der Klasse $T^+$ zu, sonst ordne das Objekt der Klasse $T^-$ zu. Die hier gewählte Temperatur (39.9 Grad Celsius) ist der entscheidende Grenzwert für die Klassifizierung. 

Allgemein, d.h. ohne explizite Angabe eines Grenzwertes kann man das mathematisch wie folgt formulieren. Sei $T^k$ die Klasse, der ein Objekt zugeordnet werden kann und $\tau_c$ der Grenzwert (hier die Grenztemperatur), den wir zu Klassifizierung der Objekte heranziehgen, dann lautet die Vorschrift:

$$
T^k = \begin{cases} 
T^- \quad \text{ wenn } \tau \leq \tau_c \\
T^+ \quad \text{ wenn } \tau > \tau_c \\
\end{cases}
\qquad \text(2)
$$


## Was ist ein Goldstandard?

In der Diagnostik bezeichnet der Begriff Goldstandard die Situation, in der der **wahren** Status der Objekte, an denen ein diagnostischer Test anwendet wird, kennen (?entspricht das schon dem Begriff supervised learning?). 

**Der Goldstandard ist also ein Verfahren, das den wahren Status der untersuchten Objekte zweifelsfrei anzeigt.**

Diese Situation ist in obigem Beispiel gegeben, und folgendend untersuchen, wie sich die Wahl des Grenzwertes auf die Zuordnung der Individuen auf die Klassen $n$ und $p$ auswirkt. In der Diagnostik wird der Grenzwert zutreffend als '*cut-off value*' bezeichnet.

Eine Funktion, die ein entsprechendes Verhalten zeigt, ist die logistische Funktion.

$$
p(\tau) = \frac{1}{1 + \exp(\beta \tau)}.
$$

Das logistische Modell hat einen Parameter. Nachfolgende Grafik zeigt die logitische Model für $\beta = -5, -3, 0, 1, 5$.

```python
def logistic(x, beta):
    return 1.0/(1.0 + np.exp(beta*x))

fig = plt.figure(figsize=(11.69,8.27))

x = np.linspace(-4,4,100)
plt.plot(x, logistic(x,-5.0), label="$\beta = -5$")
plt.plot(x, logistic(x,-3.0), label="$\beta = -3$")
plt.plot(x, logistic(x,0.0), label="$\beta = 0$")
plt.plot(x, logistic(x,3.0), label="$\beta = 3$")
plt.plot(x, logistic(x,5.0), label="$\beta = 5$")
plt.show()
```

Wir benötigen noch eine Verschiebung auf der x-Achse, um die Daten nachbilden zu können.

```python
def logistic(x, beta, alpha=0):
    return 1.0 / (1.0 + np.exp(np.dot(beta, x) + alpha))

x = np.linspace(-1,9, 100)
plt.plot(x, logistic(x,-5, 7))
plt.plot(x, logistic(x,-5,17)) 
plt.plot(x, logistic(x,-7,23))
plt.show()
```

```python
import pymc as pm

temperature = data[:,0]
D = data[:,1]

beta = pm.Normal("beta", 0., 0.001, value=0)
alpha = pm.Normal("alpha", 0, 0.001, value=0)

@pm.deterministic
def p(t=temperature, alpha=alpha, beta=beta):
    return( 1.0 / (1.0+np.exp(beta*t+alpha)))

```

```python
p.value
```

```python
observed = pm.Bernoulli("bernoulli_obs", p, value = D, observed=True)
model = pm.Model([observed, beta, alpha])

map_ = pm.MAP(model)
map_.fit()
mcmc = pm.MCMC(model)
mcmc.sample(120000, 100000, 2)
```

Wir schauen uns die Werte für $\alpha$ und $beta$ an:


```python
alpha_samples = mcmc.trace('alpha')[:, None]
beta_samples  = mcmc.trace('beta')[:, None] 

figsize(12.5,6)

plt.subplot(211)
plt.title(r"Posterior distribution of the variables $\alpha$ and $\beta$")
plt.hist(beta_samples, histtype='stepfilled', bins=35, alpha=0.85,
         label = r"posterior of $\beta$", color="#7A68A6", normed=True)
plt.legend()

plt.subplot(212)
plt.hist(alpha_samples, histtype='stepfilled', bins=35, alpha=0.85,
         label = r"posterior of $\alpha$", color= "#A60628", normed=True)
plt.legend()
plt.show()
```

##Sensitivität und Spezifität Diagnostischer Tests

Im obigen Beispiel haben wird gesehenn, dass ein diagnostischer Test untersuchte Objekte falsch und richtig klassifizieren kann. Um beurteilen zu können, ob ein diagnostischer Test richtig oder falsch klassifiziert, ist die Kenntnis der wahren Zugehörigkeit zu einer der Klassen $p$ oder $n$ notwendig. Diese Kenntnis liegt in dem oben angeführten Beispiel vor, und wir können die Sensitivität ($S_e$) und die Spezifität ($S_p$) des Testes angeben. 

Bei der Sensitivität und der Spezifität handelt es sich um bedingte Wahrscheinlichkeiten. Die Sensitivität ist die bedingte Wahrscheinlichkeit ein Objekt in die Klasse $k=p$ einzugruppieren, wenn es aus der Gruppe $k=p$ kommt, die durch entsprechende Temperaturen $T_p$ gekennzeichnet ist. Die Spezifität ist die bedingte Wahrscheinlichkeit ein Objekt in die Klasse $k=n$ einzugruppieren, wenn es aus der Gruppe $k=n$ kommt, die durch entsprechende Temperaturen $T_n$ gekennzeichnet ist. 

$$
S_e = P(k=p|T_p) \\
S_p = P(k=n|T_n) .
$$

In Abbildung (2) erkennt man, dass die Sensitivität für den gewählten Grenzwert und für die vorliegenden Daten $S_e = 1.0$ beträgt. Alle untersuchten Objekte, deren wahrer Zustand $k=p$ ist, werden durch den Klassifizierer ebenfalls der Gruppe $p$ zugeordnet. Im Beispiel wurden 100 Objekte mit wahrem Status $G_p$ und entsprechenden  Temperaturen $T_p$ erzeugt, der Klassifizierer teilt alle Individuen in die Klasse $D_p$ ein.
