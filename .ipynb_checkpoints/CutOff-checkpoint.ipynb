##2. Zuordnung
---

Im vorigen Abschnitt wurden Daten zur Klassifizierung von Objekten in die zwei Klassen "gesund" und "krank" vorgestellt. Die Klassifizierung soll anhand der Körpertemperatur erfolgen. 

Da der wahre Status der Objekte bekannt ist kann untersucht werden, wie sich die Festlegung der Entscheidungsregel auf die Einteilung der Individuen im Vergleich zum wahren Status verhält. Die Besonderheit an dem verwendeten artifiziellen Datensatz besteht darin, dass sich die Körpertemperaturen von Objekten der beiden Klassen überlappen. Die daraus resultierende Problematik soll andhand nachfolgender Grafik verdeutlicht werden.  

```python
# Daten einlesen#
# erste Spalte enthält 
data = np.genfromtxt('../data/TD.txt', 
                     delimiter='\t')
```

```python
%pylab inline
```

```python
fig, ax = plt.subplots()
#
# 1. Spalte in data sind die Temperaturen und 2. Spalte ist der wahre Zustand
ax.scatter(data[:,0], data[:,1], s=75, color='k', alpha=0.25)
ax.set_xlabel("Temperature [Deg. C]")
ax.set_ylabel("True status [0 $\equiv$ healthy]")
ax.set_yticks([0,1])
```

Es gibt Temperaturbereiche, in denen die Zuordnung zu den Klassen leicht fällt und einen Bereich, in dem die Zuordnung mit Unsicherheiten behaftet ist. Der unsichere Bereich liegt zwischen der maximalen Temperatur von Objekten $D^-$ und der minimalen Temperatur von Individuen $D^+$, also zwischen 39.831 und 40.318 Grad Celsius. 

```python
d0 = (data[:,1]==0)
d1 = (data[:,1]==1)
x_ll = np.min(data[d1,0])
x_ul = np.max(data[d0,0])
print  x_ll, x_ul
```

```python
fig, ax = plt.subplots()
#
# 1. Spalte in data sind die Temperaturen und 2. Spalte ist der wahre Zustand
ax.scatter(data[:,0], data[:,1], s= 75, color='k', alpha=0.25)
ax.set_xlabel("Temperature [Deg. C]")
ax.set_ylabel("True status [0 $\equiv$ healthy]")
ax.set_yticks([0,1])
ax.axvline(x=x_ll, color="grey")
ax.axvline(x=x_ul, color="grey")
```

Es muss aber bedacht werden, dass die Temperaturmessung zufällig aus einer Grundgesamtheit gezogen wurden. Ein andere Ziehung würde andere Werte und andere Grenzen ergeben. 

Dennoch kann man sagen, dass wir bezogen auf den vorliegenden Datensatz bei Temperaturen in der Nähe von 39.831 eher dazu tendieren, die zugehörigen Objekte der Klasse 0 zuzordenen. Bei Temperaturen in der Nähe von 40.318 tendieren wir dazu, die zugehörigen Objekte der Klasse 1 zuzuordnen.

## Lgistische Regression als binärer Klassifizierer

Die logistische Regression ist ein statistischer Klassfizierer. Für obiges Beispiel gibt die logistische Regression für jede Temperatur die Wahrscheinlichkeit der Zugehörigkeit eines Objektes mit eben dieser Temperatur zu Klasse $D^+$.

Nachfolgende führen wir die logistische Regression im Statistikprogrammpaket R aus.

```python
%load_ext rmagic
```

```python
temp = data[:,0]
```

```python
D = data[:,1]
```

```python
%%R -i temp,D -o xs,ys,obs
glm.out <- glm(D~temp, family=binomial(logit))
print(summary(glm.out))
#
# data frame erstellen für Liniengrafik
pred <- glm.out$fitted
df <- data.frame(temp=temp, pred=pred, obs=D)
df.sorted <- df[order(temp), ]
#
# Rückgabewerte erstellen
xs = df.sorted$temp
ys = df.sorted$pred
obs = df.sorted$obs
#
#einfache Grafik erstellen
plot(D~temp)
lines(xs, ys, type='l', col='blue')
abline(v=39.831, col="gray33")
abline(v=40.318, col="gray33")
abline(h=0.5, color="blue")
```
